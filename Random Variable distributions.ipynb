```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/ironhack-labs/lab-random-variable-distributions/master/files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv")
```

```python
df
```

```python
#Get the numerical variables from our dataset.
numeric_variables = df.select_dtypes(include = np.number)
numeric_variables
```

```python
#Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.

def plot_hist(numeric_variables):
    plt_size = len(numeric_variables.columns)
    fig, ax = plt.subplots(plt_size, figsize=(16,16))
    n = 0
    for i in numeric_variables.columns:
        sns.histplot( x=numeric_variables[i], ax = ax[n])
        n += 1
    
    plt.tight_layout()    
    plt.show()
```

```python
plot_hist(numeric_variables)
```

Customer Lifetime Value, Monthly Premium Auto, Number of Open Complaints follow the Exponential distribution. 

```python
#Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.
log_trans = np.log(numeric_variables)
```

```python
plot_hist(log_trans)
```
